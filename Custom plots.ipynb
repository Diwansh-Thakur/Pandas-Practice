```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series , DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize']=12,4
sb.set_style('whitegrid')
```

labeling plot features

the functional method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z=[1,2,3,4,0.5]

veh_type = ['bicycle','motorbike','car','van','stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

THe object oriented method

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60 , fontsize = 'medium')
ax.set_title('Miles per gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/galons')
```

adding a legend to your plot

the functional method

```python
plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

obeject oriented method

```python
fig = plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60 , fontsize = 'medium')
ax.set_title('Miles per gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/galons')

ax.legend(loc='best')
```

```python
fig = plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60 , fontsize = 'medium')
ax.set_title('Miles per gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/galons')

ax.set_ylim([0,45])

ax.annotate('Toyoya Corolla',xy =(19,33.9), xytext=(21,35),arrowprops=dict(facecolor='black',shrink=0.05))

ax.legend(loc='best')
```
