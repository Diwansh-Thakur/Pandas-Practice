```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize']= 5,4
```

```python
address = 'iris.data.csv'
df = pd.read_csv(filepath_or_buffer=address , header=None,sep=',')

df.columns = ['sepal length','sepal width','petal length','petal width','species']
x = df.iloc[:,0:4].values
y = df.iloc[:,4].values
df[:5]
```

Identifying outliners fron Tukey boxplot

```python
df.boxplot(return_type='dict')
plt.plot()
```

```python
sepal_width = x[:,1]

iris_outliers = (sepal_width>4)
df[iris_outliers]
```

```python
sepal_width = x[:,1]

iris_outliers = (sepal_width<2.05)
df[iris_outliers]
```

Applying Tukey outliers labelling

```python
pd.options.display.float_format = '{:.1f}'.format
x_df = pd.DataFrame(x)
print(x_df.describe())
```

Visually inspecting boxplots

```python
sns.boxplot(x='species',y='sepal length',data=df,hue='species',palette='hls',legend=False)
```

Looking at the scatterplot matrix

```python
sns.pairplot(df, hue='species',palette='hls')
```
